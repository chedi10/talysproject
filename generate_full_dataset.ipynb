## Génération bloc par bloc

On va utiliser les fonctions définies dans `generate_data.py` et les exécuter **étape par étape** :

1. Générer les clients et sauvegarder `clients.csv`
2. Générer les crédits et sauvegarder `credits.csv`
3. Générer les remboursements et sauvegarder `remboursements.csv`
4. Générer les transactions et sauvegarder `transactions.csv`
5. Générer les relations et sauvegarder `relations.csv`

Exécute chaque cellule dans l'ordre avec `Shift + Enter`.

```python
# Étape 1 — Générer et sauvegarder les clients
from generate_data import generate_clients, RAW_DIR

# Génération des clients
df_clients = generate_clients()

# Sauvegarde dans data/raw
(df_clients
 .to_csv(RAW_DIR / "clients.csv", index=False))

print("clients.csv créé dans", RAW_DIR)

df_clients.head()
```

```python
# Étape 2 — Générer et sauvegarder les crédits
from generate_data import generate_credits

# Génération des crédits à partir des clients
df_credits = generate_credits(df_clients)

# Sauvegarde dans data/raw
(df_credits
 .to_csv(RAW_DIR / "credits.csv", index=False))

print("credits.csv créé dans", RAW_DIR)

df_credits.head()
```

```python
# Étape 3 — Générer et sauvegarder les remboursements
from generate_data import generate_repayments

# Génération des remboursements à partir des crédits
df_remboursements = generate_repayments(df_credits)

# Sauvegarde dans data/raw
(df_remboursements
 .to_csv(RAW_DIR / "remboursements.csv", index=False))

print("remboursements.csv créé dans", RAW_DIR)

df_remboursements.head()
```

```python
# Étape 4 — Générer et sauvegarder les transactions
from generate_data import generate_transactions

# Génération des transactions à partir des clients
df_transactions = generate_transactions(df_clients)

# Sauvegarde dans data/raw
(df_transactions
 .to_csv(RAW_DIR / "transactions.csv", index=False))

print("transactions.csv créé dans", RAW_DIR)

df_transactions.head()
```

```python
# Étape 5 — Générer et sauvegarder les relations
from generate_data import generate_relations

# Génération des relations à partir des clients et des crédits
df_relations = generate_relations(df_clients, df_credits)

# Sauvegarde dans data/raw
(df_relations
 .to_csv(RAW_DIR / "relations.csv", index=False))

print("relations.csv créé dans", RAW_DIR)

df_relations.head()
```
